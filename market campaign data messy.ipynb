```python
# derived column like cpc and cost/conv recalculate করে save করতে হবে। 
# ***** Save করতে হয় df[] দিয়ে**

#CTR = (Clicks / Impressions) × 100 → recalculate করো
#CPC = Spend / Clicks → recalculate করো
#Conversion_Rate = (Conversions / Clicks) × 100 → recalculate করো
#Cost_Per_Conversion = Spend / Conversions → recalculate করো
#inf values (যদি থাকে) → 0 দিয়ে replace করো
```

```python

#CTR = (Clicks / Impressions) × 100 → recalculate করো
#CPC = Spend / Clicks → recalculate করো
#Conversion_Rate = (Conversions / Clicks) × 100 → recalculate করো
#Cost_Per_Conversion = Spend / Conversions → recalculate করো
#inf values (যদি থাকে) → 0 দিয়ে replace করো


# CTR recalculate: (Clicks / Impressions) * 100
df['CTR'] = round(df['Clicks'] / df['Impressions'] * 100, 2)

# CPC recalculate: Spend / Clicks
df['CPC'] = round(df['Spend'] / df['Clicks'], 2)

# Conversion_Rate recalculate: (Conversions / Clicks) * 100
df['Conversion_Rate'] = round(df['Conversions'] / df['Clicks'] * 100, 2)

# Cost_Per_Conversion recalculate: Spend / Conversions
df['Cost_Per_Conversion'] = round(df['Spend'] / df['Conversions'], 2)

# inf values থাকলে 0 দিয়ে replace করো (Conversions=0 হলে inf হয়)
df['Cost_Per_Conversion'] = df['Cost_Per_Conversion'].replace([np.inf, -np.inf], 0)

print("Derived columns recalculated!")

```

```python
# ***************************at a glance:************************
#(import pandas as pd
import numpy as np

# === LOAD DATA ===
df = pd.read_csv(r"C:\Users\anikk\Downloads\marketing_campaign_data_cleaned.csv")

# === STEP 1: Fix Negative Values ===
df['Spend'] = df['Spend'].abs()
df['CPC'] = df['CPC'].abs()
df['Cost_Per_Conversion'] = df['Cost_Per_Conversion'].abs()

# === STEP 2: Fix Extreme Outliers (IQR Method → Median Replace) ===
for col in ['Spend', 'CPC', 'Cost_Per_Conversion']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    median = df[col].median()
    df.loc[df[col] > upper, col] = median

# === STEP 3: Fill Missing Values ===
df['Channel'] = df['Channel'].fillna(df['Channel'].mode()[0])
df['Spend'] = df['Spend'].fillna(df['Spend'].median())
df['Conversions'] = df['Conversions'].fillna(df['Conversions'].median())
df['CPC'] = df['CPC'].fillna(df['CPC'].median())
df['Conversion_Rate'] = df['Conversion_Rate'].fillna(df['Conversion_Rate'].median())
df['Cost_Per_Conversion'] = df['Cost_Per_Conversion'].fillna(df['Cost_Per_Conversion'].median())

# === STEP 4: Convert Date Columns ===
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
df['End_Date'] = pd.to_datetime(df['End_Date'])

# === STEP 5: Fix Invalid Dates (Start > End → Swap) ===
mask = df['Start_Date'] > df['End_Date']
df.loc[mask, ['Start_Date', 'End_Date']] = df.loc[mask, ['End_Date', 'Start_Date']].values

# === STEP 6: Recalculate Derived Columns ===
df['CTR'] = round(df['Clicks'] / df['Impressions'] * 100, 2)
df['CPC'] = round(df['Spend'] / df['Clicks'], 2)
df['Conversion_Rate'] = round(df['Conversions'] / df['Clicks'] * 100, 2)
df['Cost_Per_Conversion'] = round(df['Spend'] / df['Conversions'], 2)
df['Cost_Per_Conversion'] = df['Cost_Per_Conversion'].replace([np.inf, -np.inf], 0)

# === STEP 7: Optimize Data Types ===
df['Active'] = df['Active'].map({'Yes': True, 'No': False})
df['Conversions'] = df['Conversions'].astype(int)

print("✅ All cleaning complete!")
df.info()
)
```
